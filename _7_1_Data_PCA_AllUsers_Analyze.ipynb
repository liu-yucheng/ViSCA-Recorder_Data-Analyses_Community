````python
import json
import os
import sys

import matplotlib.pyplot as pyplot
import numpy
import pandas
import scipy.signal as signal
import seaborn
import sklearn.decomposition as decomp
import sklearn.linear_model as linear_model
import sklearn.metrics as metrics
import sklearn.model_selection as model_selection
import sklearn.preprocessing as preproc
import statsmodels.stats.weightstats as weightstats

# %matplotlib inline

Path_Self = os.path.abspath(os.path.curdir)
Path_Input = os.path.join(Path_Self, "_7_Data_PCA_AllUsers")
Path_Output = os.path.join(Path_Self, "Figures_Computed")
Path_Output_Texts = os.path.join(Path_Self, "Texts_Computed")
os.makedirs(Path_Output_Texts, exist_ok=True)
Path_Output_Text = os.path.join(Path_Output_Texts, "_7_1_Text_PCA_AllUsers.txt")
File_Output_Text = open(Path_Output_Text, "w+")

def print_(*args, **kwargs):
    Files = [
        sys.stdout,
        File_Output_Text,
    ]

    for File_ in Files:
        kwargs["file"] = File_
        print(*args, **kwargs)
        File_.flush()
    # end for
# end def

print_(f"Begin analyzing data for all users")

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Read the preprocessed data.

Path_Input_Users = os.path.join(Path_Input, f"AllUsers_ToAnalyze.json")
Path_Input_Users_KeysToIndexes = os.path.join(Path_Input, f"AllUsers_KeysToIndexes.json")
Map_Input_Users_KeysToIndexes = {}
Path_Input_Users_IndexesToKeys = os.path.join(Path_Input, f"AllUsers_IndexesToKeys.json")
Map_Input_Users_IndexesToKeys = {}
Path_Output_Users = os.path.join(Path_Output, f"AllUsers")
os.makedirs(Path_Output_Users, exist_ok=True)

def Path_Output_Figure_Find(Path_):
    return os.path.join(Path_Output_Users, Path_)
# end def

with open(Path_Input_Users_KeysToIndexes) as File_:
    Map_Input_Users_KeysToIndexes = json.load(File_)
    Map_Input_Users_KeysToIndexes = dict(Map_Input_Users_KeysToIndexes)
# end with

with open(Path_Input_Users_IndexesToKeys) as File_:
    Map_Input_Users_IndexesToKeys = json.load(File_)
    Map_Input_Users_IndexesToKeys = dict(Map_Input_Users_IndexesToKeys)
# end with

print_(f"Begin reading file {Path_Input_Users}")
Data_ = pandas.read_json(Path_Input_Users)
print_(f"End reading file {Path_Input_Users}")

# Keep a set of selected data item key-value pairs.
Data_ = Data_[
    [
        # "Var0000", # "recorder_name",
        # "Var0001", # "timestamp.date_time_custom",
        # "Var0002", # "timestamp.date_time",
        # "Var0003", # "timestamp.unix_ms",
        # "Var0004", # "timestamp.game_time_custom",
        # "Var0005", # "timestamp.game_time",
        # "Var0006", # "timestamp.game_time_seconds",
        # "Var0007", # "timestamp.unity_fixed_delta_time_seconds",
        # "Var0008", # "timestamp.ideal_record_interval_seconds",
        # "Var0009", # "timestamp.actual_record_interval_seconds",
        # "Var0010", # "eye_gaze.spatial_pose.position.x",
        # "Var0011", # "eye_gaze.spatial_pose.position.y",
        # "Var0012", # "eye_gaze.spatial_pose.position.z",
        # "Var0013", # "eye_gaze.spatial_pose.rotation.x",
        # "Var0014", # "eye_gaze.spatial_pose.rotation.y",
        # "Var0015", # "eye_gaze.spatial_pose.rotation.z",
        "Var0016", # "eye_gaze.spatial_pose.velocity.x",
        "Var0017", # "eye_gaze.spatial_pose.velocity.y",
        "Var0018", # "eye_gaze.spatial_pose.velocity.z",
        "Var0019", # "eye_gaze.spatial_pose.angular_velocity.x",
        "Var0020", # "eye_gaze.spatial_pose.angular_velocity.y",
        "Var0021", # "eye_gaze.spatial_pose.angular_velocity.z",
        "Var0022", # "eye_gaze.spatial_pose.acceleration.x",
        "Var0023", # "eye_gaze.spatial_pose.acceleration.y",
        "Var0024", # "eye_gaze.spatial_pose.acceleration.z",
        "Var0025", # "eye_gaze.spatial_pose.angular_acceleration.x",
        "Var0026", # "eye_gaze.spatial_pose.angular_acceleration.y",
        "Var0027", # "eye_gaze.spatial_pose.angular_acceleration.z",
        # "Var0028", # "eye_gaze.spatial_pose.raw_position.x",
        # "Var0029", # "eye_gaze.spatial_pose.raw_position.y",
        # "Var0030", # "eye_gaze.spatial_pose.raw_position.z",
        # "Var0031", # "eye_gaze.spatial_pose.raw_rotation.x",
        # "Var0032", # "eye_gaze.spatial_pose.raw_rotation.y",
        # "Var0033", # "eye_gaze.spatial_pose.raw_rotation.z",
        # "Var0034", # "eye_gaze.spatial_pose.raw_velocity.x",
        # "Var0035", # "eye_gaze.spatial_pose.raw_velocity.y",
        # "Var0036", # "eye_gaze.spatial_pose.raw_velocity.z",
        # "Var0037", # "eye_gaze.spatial_pose.raw_angular_velocity.x",
        # "Var0038", # "eye_gaze.spatial_pose.raw_angular_velocity.y",
        # "Var0039", # "eye_gaze.spatial_pose.raw_angular_velocity.z",
        # "Var0040", # "eye_gaze.spatial_pose.raw_acceleration.x",
        # "Var0041", # "eye_gaze.spatial_pose.raw_acceleration.y",
        # "Var0042", # "eye_gaze.spatial_pose.raw_acceleration.z",
        # "Var0043", # "eye_gaze.spatial_pose.raw_angular_acceleration.x",
        # "Var0044", # "eye_gaze.spatial_pose.raw_angular_acceleration.y",
        # "Var0045", # "eye_gaze.spatial_pose.raw_angular_acceleration.z",
        # "Var0046", # "eye_gaze.spatial_pose.magnitude_position",
        # "Var0047", # "eye_gaze.spatial_pose.magnitude_rotation",
        # "Var0048", # "eye_gaze.spatial_pose.magnitude_velocity",
        # "Var0049", # "eye_gaze.spatial_pose.magnitude_angular_velocity",
        # "Var0050", # "eye_gaze.spatial_pose.magnitude_acceleration",
        # "Var0051", # "eye_gaze.spatial_pose.magnitude_angular_acceleration",
        # "Var0052", # "eye_gaze.spatial_pose.magnitude_raw_position",
        # "Var0053", # "eye_gaze.spatial_pose.magnitude_raw_rotation",
        # "Var0054", # "eye_gaze.spatial_pose.magnitude_raw_velocity",
        # "Var0055", # "eye_gaze.spatial_pose.magnitude_raw_angular_velocity",
        # "Var0056", # "eye_gaze.spatial_pose.magnitude_raw_acceleration",
        # "Var0057", # "eye_gaze.spatial_pose.magnitude_raw_angular_acceleration",
        # "Var0058", # "eye_gaze.viewport_pose.position.x",
        # "Var0059", # "eye_gaze.viewport_pose.position.y",
        # "Var0060", # "eye_gaze.viewport_pose.velocity.x",
        # "Var0061", # "eye_gaze.viewport_pose.velocity.y",
        # "Var0062", # "eye_gaze.viewport_pose.acceleration.x",
        # "Var0063", # "eye_gaze.viewport_pose.acceleration.y",
        # "Var0064", # "eye_gaze.viewport_pose.raw_position.x",
        # "Var0065", # "eye_gaze.viewport_pose.raw_position.y",
        # "Var0066", # "eye_gaze.viewport_pose.raw_velocity.x",
        # "Var0067", # "eye_gaze.viewport_pose.raw_velocity.y",
        # "Var0068", # "eye_gaze.viewport_pose.raw_acceleration.x",
        # "Var0069", # "eye_gaze.viewport_pose.raw_acceleration.y",
        # "Var0070", # "eye_gaze.viewport_pose.magnitude_position",
        # "Var0071", # "eye_gaze.viewport_pose.magnitude_velocity",
        # "Var0072", # "eye_gaze.viewport_pose.magnitude_acceleration",
        # # "Var0073", # "eye_gaze.viewport_pose.magnitude_raw_position",
        # "Var0074", # "eye_gaze.viewport_pose.magnitude_raw_velocity",
        # "Var0075", # "eye_gaze.viewport_pose.magnitude_raw_acceleration",
        "Var0076", # "eye_gaze.left_eye_openness",
        "Var0077", # "eye_gaze.right_eye_openness",
        "Var0078", # "face.blend_shapes_data.eyeLookDownLeft",
        "Var0079", # "face.blend_shapes_data.noseSneerLeft",
        "Var0080", # "face.blend_shapes_data.eyeLookInLeft",
        "Var0081", # "face.blend_shapes_data.browInnerUp",
        "Var0082", # "face.blend_shapes_data.browDownRight",
        "Var0083", # "face.blend_shapes_data.mouthClose",
        "Var0084", # "face.blend_shapes_data.mouthLowerDownRight",
        "Var0085", # "face.blend_shapes_data.jawOpen",
        "Var0086", # "face.blend_shapes_data.mouthUpperUpRight",
        "Var0087", # "face.blend_shapes_data.mouthShrugUpper",
        "Var0088", # "face.blend_shapes_data.mouthFunnel",
        "Var0089", # "face.blend_shapes_data.eyeLookInRight",
        "Var0090", # "face.blend_shapes_data.eyeLookDownRight",
        "Var0091", # "face.blend_shapes_data.noseSneerRight",
        "Var0092", # "face.blend_shapes_data.mouthRollUpper",
        "Var0093", # "face.blend_shapes_data.jawRight",
        "Var0094", # "face.blend_shapes_data.browDownLeft",
        "Var0095", # "face.blend_shapes_data.mouthShrugLower",
        "Var0096", # "face.blend_shapes_data.mouthRollLower",
        "Var0097", # "face.blend_shapes_data.mouthSmileLeft",
        "Var0098", # "face.blend_shapes_data.mouthPressLeft",
        "Var0099", # "face.blend_shapes_data.mouthSmileRight",
        "Var0100", # "face.blend_shapes_data.mouthPressRight",
        "Var0101", # "face.blend_shapes_data.mouthDimpleRight",
        "Var0102", # "face.blend_shapes_data.mouthLeft",
        "Var0103", # "face.blend_shapes_data.jawForward",
        "Var0104", # "face.blend_shapes_data.eyeSquintLeft",
        "Var0105", # "face.blend_shapes_data.mouthFrownLeft",
        "Var0106", # "face.blend_shapes_data.eyeBlinkLeft",
        "Var0107", # "face.blend_shapes_data.cheekSquintLeft",
        "Var0108", # "face.blend_shapes_data.browOuterUpLeft",
        "Var0109", # "face.blend_shapes_data.eyeLookUpLeft",
        "Var0110", # "face.blend_shapes_data.jawLeft",
        "Var0111", # "face.blend_shapes_data.mouthStretchLeft",
        "Var0112", # "face.blend_shapes_data.mouthPucker",
        "Var0113", # "face.blend_shapes_data.eyeLookUpRight",
        "Var0114", # "face.blend_shapes_data.browOuterUpRight",
        "Var0115", # "face.blend_shapes_data.cheekSquintRight",
        "Var0116", # "face.blend_shapes_data.eyeBlinkRight",
        "Var0117", # "face.blend_shapes_data.mouthUpperUpLeft",
        "Var0118", # "face.blend_shapes_data.mouthFrownRight",
        "Var0119", # "face.blend_shapes_data.eyeSquintRight",
        "Var0120", # "face.blend_shapes_data.mouthStretchRight",
        "Var0121", # "face.blend_shapes_data.cheekPuff",
        "Var0122", # "face.blend_shapes_data.eyeLookOutLeft",
        "Var0123", # "face.blend_shapes_data.eyeLookOutRight",
        "Var0124", # "face.blend_shapes_data.eyeWideRight",
        "Var0125", # "face.blend_shapes_data.eyeWideLeft",
        "Var0126", # "face.blend_shapes_data.mouthRight",
        "Var0127", # "face.blend_shapes_data.mouthDimpleLeft",
        "Var0128", # "face.blend_shapes_data.mouthLowerDownLeft",
        "Var0129", # "face.blend_shapes_data.tongueOut",
        # "Var0130", # "face.blend_shapes_data.viseme_PP",
        # "Var0131", # "face.blend_shapes_data.viseme_CH",
        # "Var0132", # "face.blend_shapes_data.viseme_o",
        # "Var0133", # "face.blend_shapes_data.viseme_O",
        # "Var0134", # "face.blend_shapes_data.viseme_i",
        # "Var0135", # "face.blend_shapes_data.viseme_I",
        # "Var0136", # "face.blend_shapes_data.viseme_RR",
        # "Var0137", # "face.blend_shapes_data.viseme_XX",
        # "Var0138", # "face.blend_shapes_data.viseme_aa",
        # "Var0139", # "face.blend_shapes_data.viseme_FF",
        # "Var0140", # "face.blend_shapes_data.viseme_u",
        # "Var0141", # "face.blend_shapes_data.viseme_U",
        # "Var0142", # "face.blend_shapes_data.viseme_TH",
        # "Var0143", # "face.blend_shapes_data.viseme_kk",
        # "Var0144", # "face.blend_shapes_data.viseme_SS",
        # "Var0145", # "face.blend_shapes_data.viseme_e",
        # "Var0146", # "face.blend_shapes_data.viseme_DD",
        # "Var0147", # "face.blend_shapes_data.viseme_E",
        # "Var0148", # "face.blend_shapes_data.viseme_nn",
        # "Var0149", # "face.blend_shapes_data.viseme_sil",
        "Var0150", # "face.laughing",
        # "Var0151", # "headset_spatial_pose.position.x",
        # "Var0152", # "headset_spatial_pose.position.y",
        # "Var0153", # "headset_spatial_pose.position.z",
        # "Var0154", # "headset_spatial_pose.rotation.x",
        # "Var0155", # "headset_spatial_pose.rotation.y",
        # "Var0156", # "headset_spatial_pose.rotation.z",
        "Var0157", # "headset_spatial_pose.velocity.x",
        "Var0158", # "headset_spatial_pose.velocity.y",
        "Var0159", # "headset_spatial_pose.velocity.z",
        "Var0160", # "headset_spatial_pose.angular_velocity.x",
        "Var0161", # "headset_spatial_pose.angular_velocity.y",
        "Var0162", # "headset_spatial_pose.angular_velocity.z",
        "Var0163", # "headset_spatial_pose.acceleration.x",
        "Var0164", # "headset_spatial_pose.acceleration.y",
        "Var0165", # "headset_spatial_pose.acceleration.z",
        "Var0166", # "headset_spatial_pose.angular_acceleration.x",
        "Var0167", # "headset_spatial_pose.angular_acceleration.y",
        "Var0168", # "headset_spatial_pose.angular_acceleration.z",
        # "Var0169", # "headset_spatial_pose.raw_position.x",
        # "Var0170", # "headset_spatial_pose.raw_position.y",
        # "Var0171", # "headset_spatial_pose.raw_position.z",
        # "Var0172", # "headset_spatial_pose.raw_rotation.x",
        # "Var0173", # "headset_spatial_pose.raw_rotation.y",
        # "Var0174", # "headset_spatial_pose.raw_rotation.z",
        # "Var0175", # "headset_spatial_pose.raw_velocity.x",
        # "Var0176", # "headset_spatial_pose.raw_velocity.y",
        # "Var0177", # "headset_spatial_pose.raw_velocity.z",
        # "Var0178", # "headset_spatial_pose.raw_angular_velocity.x",
        # "Var0179", # "headset_spatial_pose.raw_angular_velocity.y",
        # "Var0180", # "headset_spatial_pose.raw_angular_velocity.z",
        # "Var0181", # "headset_spatial_pose.raw_acceleration.x",
        # "Var0182", # "headset_spatial_pose.raw_acceleration.y",
        # "Var0183", # "headset_spatial_pose.raw_acceleration.z",
        # "Var0184", # "headset_spatial_pose.raw_angular_acceleration.x",
        # "Var0185", # "headset_spatial_pose.raw_angular_acceleration.y",
        # "Var0186", # "headset_spatial_pose.raw_angular_acceleration.z",
        # "Var0187", # "headset_spatial_pose.magnitude_position",
        # "Var0188", # "headset_spatial_pose.magnitude_rotation",
        # "Var0189", # "headset_spatial_pose.magnitude_velocity",
        # "Var0190", # "headset_spatial_pose.magnitude_angular_velocity",
        # "Var0191", # "headset_spatial_pose.magnitude_acceleration",
        # "Var0192", # "headset_spatial_pose.magnitude_angular_acceleration",
        # "Var0193", # "headset_spatial_pose.magnitude_raw_position",
        # "Var0194", # "headset_spatial_pose.magnitude_raw_rotation",
        # "Var0195", # "headset_spatial_pose.magnitude_raw_velocity",
        # "Var0196", # "headset_spatial_pose.magnitude_raw_angular_velocity",
        # "Var0197", # "headset_spatial_pose.magnitude_raw_acceleration",
        # "Var0198", # "headset_spatial_pose.magnitude_raw_angular_acceleration",
        # "Var0199", # "left_controller_spatial_pose.position.x",
        # "Var0200", # "left_controller_spatial_pose.position.y",
        # "Var0201", # "left_controller_spatial_pose.position.z",
        # "Var0202", # "left_controller_spatial_pose.rotation.x",
        # "Var0203", # "left_controller_spatial_pose.rotation.y",
        # "Var0204", # "left_controller_spatial_pose.rotation.z",
        # "Var0205", # "left_controller_spatial_pose.velocity.x",
        # "Var0206", # "left_controller_spatial_pose.velocity.y",
        # "Var0207", # "left_controller_spatial_pose.velocity.z",
        # "Var0208", # "left_controller_spatial_pose.angular_velocity.x",
        # "Var0209", # "left_controller_spatial_pose.angular_velocity.y",
        # "Var0210", # "left_controller_spatial_pose.angular_velocity.z",
        # "Var0211", # "left_controller_spatial_pose.acceleration.x",
        # "Var0212", # "left_controller_spatial_pose.acceleration.y",
        # "Var0213", # "left_controller_spatial_pose.acceleration.z",
        # "Var0214", # "left_controller_spatial_pose.angular_acceleration.x",
        # "Var0215", # "left_controller_spatial_pose.angular_acceleration.y",
        # "Var0216", # "left_controller_spatial_pose.angular_acceleration.z",
        # "Var0217", # "left_controller_spatial_pose.raw_position.x",
        # "Var0218", # "left_controller_spatial_pose.raw_position.y",
        # "Var0219", # "left_controller_spatial_pose.raw_position.z",
        # "Var0220", # "left_controller_spatial_pose.raw_rotation.x",
        # "Var0221", # "left_controller_spatial_pose.raw_rotation.y",
        # "Var0222", # "left_controller_spatial_pose.raw_rotation.z",
        # "Var0223", # "left_controller_spatial_pose.raw_velocity.x",
        # "Var0224", # "left_controller_spatial_pose.raw_velocity.y",
        # "Var0225", # "left_controller_spatial_pose.raw_velocity.z",
        # "Var0226", # "left_controller_spatial_pose.raw_angular_velocity.x",
        # "Var0227", # "left_controller_spatial_pose.raw_angular_velocity.y",
        # "Var0228", # "left_controller_spatial_pose.raw_angular_velocity.z",
        # "Var0229", # "left_controller_spatial_pose.raw_acceleration.x",
        # "Var0230", # "left_controller_spatial_pose.raw_acceleration.y",
        # "Var0231", # "left_controller_spatial_pose.raw_acceleration.z",
        # "Var0232", # "left_controller_spatial_pose.raw_angular_acceleration.x",
        # "Var0233", # "left_controller_spatial_pose.raw_angular_acceleration.y",
        # "Var0234", # "left_controller_spatial_pose.raw_angular_acceleration.z",
        # "Var0235", # "left_controller_spatial_pose.magnitude_position",
        # "Var0236", # "left_controller_spatial_pose.magnitude_rotation",
        # "Var0237", # "left_controller_spatial_pose.magnitude_velocity",
        # "Var0238", # "left_controller_spatial_pose.magnitude_angular_velocity",
        # "Var0239", # "left_controller_spatial_pose.magnitude_acceleration",
        # "Var0240", # "left_controller_spatial_pose.magnitude_angular_acceleration",
        # "Var0241", # "left_controller_spatial_pose.magnitude_raw_position",
        # "Var0242", # "left_controller_spatial_pose.magnitude_raw_rotation",
        # "Var0243", # "left_controller_spatial_pose.magnitude_raw_velocity",
        # "Var0244", # "left_controller_spatial_pose.magnitude_raw_angular_velocity",
        # "Var0245", # "left_controller_spatial_pose.magnitude_raw_acceleration",
        # "Var0246", # "left_controller_spatial_pose.magnitude_raw_angular_acceleration",
        # "Var0247", # "right_controller_spatial_pose.position.x",
        # "Var0248", # "right_controller_spatial_pose.position.y",
        # "Var0249", # "right_controller_spatial_pose.position.z",
        # "Var0250", # "right_controller_spatial_pose.rotation.x",
        # "Var0251", # "right_controller_spatial_pose.rotation.y",
        # "Var0252", # "right_controller_spatial_pose.rotation.z",
        # "Var0253", # "right_controller_spatial_pose.velocity.x",
        # "Var0254", # "right_controller_spatial_pose.velocity.y",
        # "Var0255", # "right_controller_spatial_pose.velocity.z",
        # "Var0256", # "right_controller_spatial_pose.angular_velocity.x",
        # "Var0257", # "right_controller_spatial_pose.angular_velocity.y",
        # "Var0258", # "right_controller_spatial_pose.angular_velocity.z",
        # "Var0259", # "right_controller_spatial_pose.acceleration.x",
        # "Var0260", # "right_controller_spatial_pose.acceleration.y",
        # "Var0261", # "right_controller_spatial_pose.acceleration.z",
        # "Var0262", # "right_controller_spatial_pose.angular_acceleration.x",
        # "Var0263", # "right_controller_spatial_pose.angular_acceleration.y",
        # "Var0264", # "right_controller_spatial_pose.angular_acceleration.z",
        # "Var0265", # "right_controller_spatial_pose.raw_position.x",
        # "Var0266", # "right_controller_spatial_pose.raw_position.y",
        # "Var0267", # "right_controller_spatial_pose.raw_position.z",
        # "Var0268", # "right_controller_spatial_pose.raw_rotation.x",
        # "Var0269", # "right_controller_spatial_pose.raw_rotation.y",
        # "Var0270", # "right_controller_spatial_pose.raw_rotation.z",
        # "Var0271", # "right_controller_spatial_pose.raw_velocity.x",
        # "Var0272", # "right_controller_spatial_pose.raw_velocity.y",
        # "Var0273", # "right_controller_spatial_pose.raw_velocity.z",
        # "Var0274", # "right_controller_spatial_pose.raw_angular_velocity.x",
        # "Var0275", # "right_controller_spatial_pose.raw_angular_velocity.y",
        # "Var0276", # "right_controller_spatial_pose.raw_angular_velocity.z",
        # "Var0277", # "right_controller_spatial_pose.raw_acceleration.x",
        # "Var0278", # "right_controller_spatial_pose.raw_acceleration.y",
        # "Var0279", # "right_controller_spatial_pose.raw_acceleration.z",
        # "Var0280", # "right_controller_spatial_pose.raw_angular_acceleration.x",
        # "Var0281", # "right_controller_spatial_pose.raw_angular_acceleration.y",
        # "Var0282", # "right_controller_spatial_pose.raw_angular_acceleration.z",
        # "Var0283", # "right_controller_spatial_pose.magnitude_position",
        # "Var0284", # "right_controller_spatial_pose.magnitude_rotation",
        # "Var0285", # "right_controller_spatial_pose.magnitude_velocity",
        # "Var0286", # "right_controller_spatial_pose.magnitude_angular_velocity",
        # "Var0287", # "right_controller_spatial_pose.magnitude_acceleration",
        # "Var0288", # "right_controller_spatial_pose.magnitude_angular_acceleration",
        # "Var0289", # "right_controller_spatial_pose.magnitude_raw_position",
        # "Var0290", # "right_controller_spatial_pose.magnitude_raw_rotation",
        # "Var0291", # "right_controller_spatial_pose.magnitude_raw_velocity",
        # "Var0292", # "right_controller_spatial_pose.magnitude_raw_angular_velocity",
        # "Var0293", # "right_controller_spatial_pose.magnitude_raw_acceleration",
        # "Var0294", # "right_controller_spatial_pose.magnitude_raw_angular_acceleration",
        # "Var0295", # "game_configs.user_index",
        # "Var0296", # "game_configs.preset_index",
        # "Var0297", # "game_configs.vehicle_opacity",
        # "Var0298", # "game_configs.autopilot_enabled",
        # "Var0299", # "game_configs.random_seed_terrain",
        # "Var0300", # "game_configs.random_seed_autopilot",
        # "Var0301", # "game_configs.random_seed_specials",
        # "Var0302", # "game_configs.specials_index",
        # "Var0303", # "game_configs.specials.color1_path.r",
        # "Var0304", # "game_configs.specials.color1_path.g",
        # "Var0305", # "game_configs.specials.color1_path.b",
        # "Var0306", # "game_configs.specials.color1_path.a",
        # "Var0307", # "game_configs.specials.color2_not_path.r",
        # "Var0308", # "game_configs.specials.color2_not_path.g",
        # "Var0309", # "game_configs.specials.color2_not_path.b",
        # "Var0310", # "game_configs.specials.color2_not_path.a",
        # "Var0311", # "game_configs.specials.color3_not_path.r",
        # "Var0312", # "game_configs.specials.color3_not_path.g",
        # "Var0313", # "game_configs.specials.color3_not_path.b",
        # "Var0314", # "game_configs.specials.color3_not_path.a",
        # "Var0315", # "game_configs.specials.color1_name_path",
        # "Var0316", # "game_configs.specials.color2_name_not_path",
        # "Var0317", # "game_configs.specials.color3_name_not_path",
        # "Var0318", # "game_play.scene_name",
        # "Var0319", # "game_play.autopilot_enabled",
        # "Var0320", # "game_play.locomotion.left_joystick_input.x",
        # "Var0321", # "game_play.locomotion.left_joystick_input.y",
        # "Var0322", # "game_play.locomotion.right_joystick_input.x",
        # "Var0323", # "game_play.locomotion.right_joystick_input.y",
        # "Var0324", # "game_play.locomotion.combined_input.x",
        # "Var0325", # "game_play.locomotion.combined_input.y",
        # "Var0326", # "game_play.locomotion.active_input.enabled",
        # "Var0327", # "game_play.locomotion.active_input.input_value.x",
        # "Var0328", # "game_play.locomotion.active_input.input_value.y",
        # "Var0329", # "game_play.locomotion.passive_input.enabled",
        # "Var0330", # "game_play.locomotion.passive_input.input_value.x",
        # "Var0331", # "game_play.locomotion.passive_input.input_value.y",
        # "Var0332", # "game_play.locomotion.spatial_pose.position.x",
        # "Var0333", # "game_play.locomotion.spatial_pose.position.y",
        # "Var0334", # "game_play.locomotion.spatial_pose.position.z",
        # "Var0335", # "game_play.locomotion.spatial_pose.rotation.x",
        # "Var0336", # "game_play.locomotion.spatial_pose.rotation.y",
        # "Var0337", # "game_play.locomotion.spatial_pose.rotation.z",
        "Var0338", # "game_play.locomotion.spatial_pose.velocity.x",
        "Var0339", # "game_play.locomotion.spatial_pose.velocity.y",
        "Var0340", # "game_play.locomotion.spatial_pose.velocity.z",
        "Var0341", # "game_play.locomotion.spatial_pose.angular_velocity.x",
        "Var0342", # "game_play.locomotion.spatial_pose.angular_velocity.y",
        "Var0343", # "game_play.locomotion.spatial_pose.angular_velocity.z",
        "Var0344", # "game_play.locomotion.spatial_pose.acceleration.x",
        "Var0345", # "game_play.locomotion.spatial_pose.acceleration.y",
        "Var0346", # "game_play.locomotion.spatial_pose.acceleration.z",
        "Var0347", # "game_play.locomotion.spatial_pose.angular_acceleration.x",
        "Var0348", # "game_play.locomotion.spatial_pose.angular_acceleration.y",
        "Var0349", # "game_play.locomotion.spatial_pose.angular_acceleration.z",
        # "Var0350", # "game_play.locomotion.spatial_pose.raw_position.x",
        # "Var0351", # "game_play.locomotion.spatial_pose.raw_position.y",
        # "Var0352", # "game_play.locomotion.spatial_pose.raw_position.z",
        # "Var0353", # "game_play.locomotion.spatial_pose.raw_rotation.x",
        # "Var0354", # "game_play.locomotion.spatial_pose.raw_rotation.y",
        # "Var0355", # "game_play.locomotion.spatial_pose.raw_rotation.z",
        # "Var0356", # "game_play.locomotion.spatial_pose.raw_velocity.x",
        # "Var0357", # "game_play.locomotion.spatial_pose.raw_velocity.y",
        # "Var0358", # "game_play.locomotion.spatial_pose.raw_velocity.z",
        # "Var0359", # "game_play.locomotion.spatial_pose.raw_angular_velocity.x",
        # "Var0360", # "game_play.locomotion.spatial_pose.raw_angular_velocity.y",
        # "Var0361", # "game_play.locomotion.spatial_pose.raw_angular_velocity.z",
        # "Var0362", # "game_play.locomotion.spatial_pose.raw_acceleration.x",
        # "Var0363", # "game_play.locomotion.spatial_pose.raw_acceleration.y",
        # "Var0364", # "game_play.locomotion.spatial_pose.raw_acceleration.z",
        # "Var0365", # "game_play.locomotion.spatial_pose.raw_angular_acceleration.x",
        # "Var0366", # "game_play.locomotion.spatial_pose.raw_angular_acceleration.y",
        # "Var0367", # "game_play.locomotion.spatial_pose.raw_angular_acceleration.z",
        # "Var0368", # "game_play.locomotion.spatial_pose.magnitude_position",
        # "Var0369", # "game_play.locomotion.spatial_pose.magnitude_rotation",
        # "Var0370", # "game_play.locomotion.spatial_pose.magnitude_velocity",
        # "Var0371", # "game_play.locomotion.spatial_pose.magnitude_angular_velocity",
        # "Var0372", # "game_play.locomotion.spatial_pose.magnitude_acceleration",
        # "Var0373", # "game_play.locomotion.spatial_pose.magnitude_angular_acceleration",
        # "Var0374", # "game_play.locomotion.spatial_pose.magnitude_raw_position",
        # "Var0375", # "game_play.locomotion.spatial_pose.magnitude_raw_rotation",
        # "Var0376", # "game_play.locomotion.spatial_pose.magnitude_raw_velocity",
        # "Var0377", # "game_play.locomotion.spatial_pose.magnitude_raw_angular_velocity",
        # "Var0378", # "game_play.locomotion.spatial_pose.magnitude_raw_acceleration",
        # "Var0379", # "game_play.locomotion.spatial_pose.magnitude_raw_angular_acceleration",
        # "Var0380", # "sickness.button_input.a",
        # "Var0381", # "sickness.button_input.b",
        # "Var0382", # "sickness.button_input.x",
        # "Var0383", # "sickness.button_input.y",
        # "Var0384", # "sickness.reported",
        "Var0385", # "sickness.deduced",
        # "Var0386", # "sickness.predicted",
        # "Var0387", # "continuous_ema.time_interval_ideal",
        # "Var0388", # "continuous_ema.time_interval_actual",
        # "Var0389", # "continuous_ema.exponent_rectification",
        # "Var0390", # "continuous_ema.alpha_ideal",
        # "Var0391", # "continuous_ema.alpha_actual",
        # "Var0392", # "continuous_ema._1_minus_alpha_ideal",
        # "Var0393", # "continuous_ema._1_minus_alpha_actual",
    ]
]

print_(
    "Begin Data_\n"
    + Data_.describe().__str__()
    + "\nEnd Data_"
)

print_("\n\n")

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Compute textual PCA cumulative explained variance ratio info.

Data_Indep = Data_[
    Data_.columns.difference(
        [
            "Var0385",  # "sickness.deduced",
        ]
    )
]

Data_Dep = Data_[
    [
        "Var0385",  # "sickness.deduced",
    ]
]

StandardScaler_ = preproc.StandardScaler()
Data_Indep_Norm = StandardScaler_.fit_transform(Data_Indep)
Data_Dep_Norm = StandardScaler_.fit_transform(Data_Dep)
PCA_ = decomp.PCA()
PCA_Fit = PCA_.fit(Data_Indep_Norm, Data_Dep)
PCA_FitTransform = PCA_.fit_transform(Data_Indep_Norm, Data_Dep)
PCs_Features_Count = PCA_Fit.n_features_in_
PCs_Names = [f"PC{Index:04d}" for Index in range(0, PCs_Features_Count)]
ExplVarRatios = PCA_Fit.explained_variance_ratio_
ExplVarRatios_Cumul = numpy.cumsum(ExplVarRatios)

print_(
    f"<<<< All Users' PCA Cumulative Explained Variance Ratios >>>>\n"
    + "PC-Name\t\tCumulative-Explained-Variance-Ratio"
)

for PC_Name, ExplVarRatio_Cumul in zip(PCs_Names, ExplVarRatios_Cumul):
    print_(f"{PC_Name}\t\t{ExplVarRatio_Cumul}")
# end for

print_("\n\n")

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Compute graphical PCA cumulative explained variance ratio info.

ColorMap = seaborn.hls_palette(
    n_colors=6,
    h=0.92,
    l=0.50,
    s=0.25,
)

Palette_ = seaborn.color_palette(ColorMap)
Palette_.reverse()

seaborn.set_theme(
    context="notebook",
    style="whitegrid",
    palette=Palette_,
    font="monospace",
)

Data_ExplVarRatios_Cumul = pandas.DataFrame({
    "PC-Index": list(range(1, PCs_Features_Count + 1)),
    "PC-Name": PCs_Names,
    "Cumulative-Explained-Variance-Ratio": ExplVarRatios_Cumul,
})

Plot = seaborn.relplot(
    Data_ExplVarRatios_Cumul,
    x="PC-Index",
    y="Cumulative-Explained-Variance-Ratio",
    kind="line",
)

Plot.figure.suptitle(f"All Users' PCA Cumulative Explained Variance Ratios")
pyplot.tight_layout(rect=[0, 0, 1, 1])
pyplot.show()

Plot.figure.savefig(
    Path_Output_Figure_Find(f"fig_AllUsers_PCA_ExplVarRatios_Cumul.png"),
    dpi=Plot.figure.dpi,
    bbox_inches="tight",
)

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Compute textual descriptive statistics of PCA weights.

PCs_Weights = PCA_Fit.components_

Data_Weights = pandas.DataFrame.from_dict(
    dict(
        zip(PCs_Names, PCs_Weights)
    )
)

Data_Weights["Variable-Index"] = Data_Indep.columns.values
Data_Weights = Data_Weights.set_index("Variable-Index")
pandas.set_option('display.max_rows', None)
pandas.set_option('display.max_columns', None)

print_(
    f"<<<< All Users' PCA Weights Descriptive Statistics >>>>\n"
    + Data_Weights.describe().__str__()
)

pandas.reset_option('display.max_rows')
pandas.reset_option('display.max_columns')
print_("\n\n")

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Compute textual PCA weights info.

pandas.set_option('display.max_rows', None)
pandas.set_option('display.max_columns', None)

print_(
    f"<<<< All Users' PCA Weights >>>>\n"
    + Data_Weights.__str__()
)

pandas.reset_option('display.max_rows')
pandas.reset_option('display.max_columns')
print_("\n\n")

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Compute graphical heatmap of PCA weights.

Plot = seaborn.heatmap(
    Data_Weights,
    cmap="Spectral",
)

Plot.figure.suptitle(f"All Users' PCA Weights Heatmap")
pyplot.xlabel("Principal-Component")
pyplot.tight_layout(rect=[0, 0, 1, 1])
pyplot.show()

Plot.figure.savefig(
    Path_Output_Figure_Find(f"fig_AllUsers_PCA_WeightsHeatmap.png"),
    dpi=Plot.figure.dpi,
    bbox_inches="tight",
)

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Compute textual independent variable PCA non-trivial weight counts statistics.

Indexes_ToNontrivWeightCounts = {}

for Index_PC, Vars_Weights in enumerate(PCs_Weights):
    Data_Vars_Weights = pandas.DataFrame(Vars_Weights)
    Vars_Weights_Mean = Data_Vars_Weights.mean()
    Vars_Weights_Mean = float(Vars_Weights_Mean.iloc[0])
    Vars_Weights_StdVar = Data_Vars_Weights.std()
    Vars_Weights_StdVar = float(Vars_Weights_StdVar.iloc[0])

    for Index_Var_Indep, Var_Weight in enumerate(Vars_Weights):
        ExplVarRatio_Cumul = ExplVarRatios_Cumul[Index_PC]

        if \
            ExplVarRatio_Cumul <= 0.9 \
            and numpy.abs(Var_Weight - Vars_Weights_Mean) >= 1.65 * Vars_Weights_StdVar \
        :
            Name_Var = Data_Indep.columns.values[Index_Var_Indep]

            if Name_Var not in Indexes_ToNontrivWeightCounts:
                Indexes_ToNontrivWeightCounts[Name_Var] = 0
            # end if

            Indexes_ToNontrivWeightCounts[Name_Var] += 1
        # end if
    # end for
# end for

NontrivWeightCounts_AndIndexes_AndKeys = []

for Index_Var in Indexes_ToNontrivWeightCounts.keys():
    NontrivWeightCounts_AndIndexes_AndKeys.append(
        (
            Indexes_ToNontrivWeightCounts[Index_Var],
            Index_Var,
            Map_Input_Users_IndexesToKeys[Index_Var],
        )
    )
# end for

NontrivWeightCounts_AndIndexes_AndKeys.sort()
NontrivWeightCounts_AndIndexes_AndKeys.reverse()

print_(
    f"<<<< All Users' Independent Variable Non-trivial Weight Counts >>>>\n"
    + "Rank\tVariable-Index\tCount\tVariable-Key"
)

NonTrivWeightCounts_PrevValue = NontrivWeightCounts_AndIndexes_AndKeys[0]
NonTrivWeightCounts_CurrRank = 1

for Index, (Count, Var_Index, Var_Key) in enumerate(NontrivWeightCounts_AndIndexes_AndKeys):
    if Count != NonTrivWeightCounts_PrevValue:
        NonTrivWeightCounts_CurrRank = Index + 1
    # end if

    print_(f"{NonTrivWeightCounts_CurrRank}\t{Var_Index}\t\t{Count}\t{Var_Key}")
    NonTrivWeightCounts_PrevValue = Count
# end for

print_("\n\n")

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Compute graphical independent variable PCA non-trivial weight counts statistics.

Data_NontrivialWeightCounts = pandas.DataFrame({
    "Variable-Index": [Index for Count, Index, Key in NontrivWeightCounts_AndIndexes_AndKeys][0:16],
    "Nontrivial-Weight-Count": [Count for Count, Index, Key in NontrivWeightCounts_AndIndexes_AndKeys][0:16],
})

Plot = seaborn.catplot(
    Data_NontrivialWeightCounts,
    x="Variable-Index",
    y="Nontrivial-Weight-Count",
    kind="bar"
)

Plot.ax.tick_params("x", labelrotation=90.0)
Plot.figure.suptitle(f"All Users' Independent Variable PCA Non-trivial Weight Counts")
pyplot.tight_layout(rect=[0, 0, 1, 1])
pyplot.show()

Plot.figure.savefig(
    Path_Output_Figure_Find(f"fig_AllUsers_PCA_IndepVar_NontrivialWeightCounts.png"),
    dpi=Plot.figure.dpi,
    bbox_inches="tight",
)

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Complete textual PCA (all dimensions) VR sickness prediction info.

Full_DimenCount = len(ExplVarRatios_Cumul)
PCA_LowDimen = decomp.PCA(Full_DimenCount)
Full_Indep = PCA_LowDimen.fit_transform(Data_Indep_Norm)
Full_Dep = Data_Dep

Train_Indep, Test_Indep, Train_Dep, Test_Dep \
= model_selection.train_test_split(
    Full_Indep,
    Full_Dep,
    test_size=0.2,
    random_state=0
)

RegreModel = linear_model.LogisticRegression(
    penalty="elasticnet",
    tol=1e-2,
    random_state=0,
    solver="saga",
    max_iter=100,
    l1_ratio=0.5,
)

RegreModel.fit(Train_Indep, Train_Dep)
Test_Dep_Predict = RegreModel.predict(Test_Indep)

print_(
    f"<<<< All Users' PCA VR Sickness Prediction Results >>>>\n"
    + f"Predictor Model Type:\tLogistic Regression\n"
    + f"PCS Space Type:\t\tAll Dimensions\n"
    + f"PCA Dimension Count:\t{Full_DimenCount}\n"
    + metrics.classification_report(Test_Dep, Test_Dep_Predict).__str__()
)

print_("\n\n")

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Complete graphical PCA (all dimensions) VR sickness prediction info.

ConfuseMatrix = metrics.confusion_matrix(Test_Dep, Test_Dep_Predict)

Plot = seaborn.heatmap(
    ConfuseMatrix,
    cmap="Spectral",
    annot=True,
    fmt="d",
)

Plot.figure.suptitle(
    f"All Users' PCA (All Dimensions)\n"
    + f"VR Sickness Prediction Confusion Matrix"
)

pyplot.xlabel("Predicted VR Sickness")
pyplot.ylabel("Actual VR Sickness")
pyplot.tight_layout(rect=[0, 0, 1, 1])
pyplot.show()

Plot.figure.savefig(
    Path_Output_Figure_Find(f"fig_AllUsers_PCA_AllDimen_ConfuseMatrix.png"),
    dpi=Plot.figure.dpi,
    bbox_inches="tight",
)

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Complete textual PCA (90% variance explained) VR sickness prediction info.

Full_DimenCount = 1

for Index_PC, ExplVarRatio_Cumul in enumerate(ExplVarRatios_Cumul):
    if ExplVarRatio_Cumul >= 0.9:
        Full_DimenCount = Index_PC + 1
        break
    else:
        Full_DimenCount += 1
    # end if
# end for

PCA_LowDimen = decomp.PCA(Full_DimenCount)
Full_Indep = PCA_LowDimen.fit_transform(Data_Indep_Norm)
Full_Dep = Data_Dep

Train_Indep, Test_Indep, Train_Dep, Test_Dep \
= model_selection.train_test_split(
    Full_Indep,
    Full_Dep,
    test_size=0.2,
    random_state=0
)

RegreModel = linear_model.LogisticRegression(
    penalty="elasticnet",
    tol=1e-2,
    random_state=0,
    solver="saga",
    max_iter=100,
    l1_ratio=0.5,
)

RegreModel.fit(Train_Indep, Train_Dep)
Test_Dep_Predict = RegreModel.predict(Test_Indep)

print_(
    f"<<<< All Users' PCA VR Sickness Prediction Results >>>>\n"
    + f"Predictor Model Type:\tLogistic Regression\n"
    + f"PCS Space Type:\t\t90% Variance Explained\n"
    + f"PCA Dimension Count:\t{Full_DimenCount}\n"
    + metrics.classification_report(Test_Dep, Test_Dep_Predict).__str__()
)

print_("\n\n")

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Complete graphical PCA (90% variance explained) VR sickness prediction info.

ConfuseMatrix = metrics.confusion_matrix(Test_Dep, Test_Dep_Predict)

Plot = seaborn.heatmap(
    ConfuseMatrix,
    cmap="Spectral",
    annot=True,
    fmt="d",
)

Plot.figure.suptitle(
    f"All Users' PCA (90% Variance Explained)\n"
    + f"VR Sickness Prediction Confusion Matrix"
)

pyplot.xlabel("Predicted VR Sickness")
pyplot.ylabel("Actual VR Sickness")
pyplot.tight_layout(rect=[0, 0, 1, 1])
pyplot.show()

Plot.figure.savefig(
    Path_Output_Figure_Find(f"fig_AllUsers_PCA_0.9VarExpl_ConfuseMatrix.png"),
    dpi=Plot.figure.dpi,
    bbox_inches="tight",
)

# ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
# - Complete analyzing the preprocessed data.

print_(f"End analyzing data for all users")
print_("\n\n\n\n")

File_Output_Text.close()
````
