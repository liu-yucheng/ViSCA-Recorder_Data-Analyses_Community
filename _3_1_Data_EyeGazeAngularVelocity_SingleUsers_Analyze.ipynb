# 3.1. Analyzing Single Users' Eye-gaze Angular Velocity Data

Analyzing single users' VR sickness vs. eye-gaze angular velocity data.

Note the following items.

- We need to run `./_0_1_Data_Shared_Prepare.ipynb` before running the below code blocks.
- Then, we need to run `./_3_0_Data_EyeGazeAngularVelocity_SingleUsers_Prepare.ipynb` before running the below code blocks.

Read the following data files to analyze from the `./_3_Data_EyeGazeAngularVelocity_SingleUsers` folder.

- `User<Index_User>_ToAnalyze.json`.

Preprocess the following data files to analyze from the `./_3_Data_EyeGazeAngularVelocity_SingleUsers` folder.

- `User<Index_User>_ToAnalyze.json`.
  - Create a copy of column `"Sickness"` called `"SicknessCategory"` of type `category`.

Analyze the preprocessed data by going through the following steps.

- Read the preprocessed data.
- Compute textual descriptive statistics of `EyeGazeAngularVelocity`.
- Compute a graphical category plot of the `EyeGazeAngularVelocity` data.
- Compute textual descriptive statistics of the `EyeGazeAngularVelocity` vs `SicknessCategory == 0` data.
- Compute textual descriptive statistics of the `EyeGazeAngularVelocity` vs `SicknessCategory == 1` data.
- Compute textual summary of compare means statistical tests on `"EyeGazeAngularVelocity"` items with `SicknessCategory == "0"` vs `SicknessCategory == "1"`.
- Compute a graphical category plot of the `EyeGazeAngularVelocity` vs `SicknessCategory` data.
- Compute a graphical line plot of the `Time` vs. `EyeGazeAngularVelocity` comprehensive (with-sickness and without-sickness) data.
- Compute a graphical line plot of the `Time` vs. `EyeGazeAngularVelocity` with-sickness data.
- Compute a graphical line plot of the `Time` vs. `EyeGazeAngularVelocity` without-sickness data.
- Compute textual and graphical least square spectual analyses info of the `Time` vs. `EyeGazeAngularVelocity` comprehensive (with-sickness and without-sickness) data. We can deem the least square spectual analyses as a kind of discrete Fourier transform analyses designed for time-series data with uneven time-steps.
- Compute textual and graphical least square spectual analyses info of the `Time` vs. `EyeGazeAngularVelocity` with-sickness data. We can deem the least square spectual analyses as a kind of discrete Fourier transform analyses designed for time-series data with uneven time-steps. 
- Compute textual and graphical least square spectual analyses info of the `Time` vs. `EyeGazeAngularVelocity` without-sickness data. We can deem the least square spectual analyses as a kind of discrete Fourier transform analyses designed for time-series data with uneven time-steps.
- Compute textual and graphical info of the differences between the comprehensive and with-sickness frequency spectrums.
- Compute textual and graphical info of the differences between the with-sickness and without-sickness frequency spectrums.
- Complete analyzing the preprocessed data.

````python
import json
import os
import sys

import numpy
import pandas
import scipy.signal as signal
import statsmodels.stats.weightstats as StatsModels_WeightStats
import seaborn
import matplotlib.pyplot as pyplot

# %matplotlib inline

Path_Self = os.path.abspath(os.path.curdir)
Path_Input = os.path.join(Path_Self, "_3_Data_EyeGazeAngularVelocity_SingleUsers")
Path_Output = os.path.join(Path_Self, "Figures_Computed")
Path_Output_Texts = os.path.join(Path_Self, "Texts_Computed")
os.makedirs(Path_Output_Texts, exist_ok=True)
Path_Output_Text = os.path.join(Path_Output_Texts, "_3_1_Text_EyeGazeAngularVelocity_SingleUsers.txt")
File_Output_Text = open(Path_Output_Text, "w+")

def print_(*args, **kwargs):
    Files = [
        sys.stdout,
        File_Output_Text,
    ]

    for File_ in Files:
        kwargs["file"] = File_
        print(*args, **kwargs)
        File_.flush()
    # end for
# end def

Indexes_Users = list(range(0, 4, 1))

for Index_User in Indexes_Users:
    print_(f"Begin analyzing data for user {Index_User}")

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Read the preprocessed data.

    Path_Input_Users = os.path.join(Path_Input, f"User{Index_User}_ToAnalyze.json")
    Path_Output_Users = os.path.join(Path_Output, f"User{Index_User}")
    os.makedirs(Path_Output_Users, exist_ok=True)

    def Path_Output_Figure_Find(Path_):
        return os.path.join(Path_Output_Users, Path_)
    # end def

    Data_ = pandas.read_json(Path_Input_Users)
    Data_["SicknessCategory"] = Data_["Sickness"].astype("category")

    print_(
        "Begin Data_\n"
        + Data_.describe().__str__()
        + "\nEnd Data_"
    )

    Data_WithSickness = Data_[Data_["SicknessCategory"] == 1]

    print_(
        "Begin Data_WithSickness\n"
        + Data_WithSickness.describe().__str__()
        + "\nEnd Data_WithSickness"
    )

    Data_WithoutSickness = Data_[Data_["SicknessCategory"] == 0]

    print_(
        "Begin Data_WithoutSickness\n"
        + Data_WithoutSickness.describe().__str__()
        + "\nEnd Data_WithoutSickness"
    )

    print_("\n\n")

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute textual descriptive statistics of `EyeGazeAngularVelocity`.

    print_(
        "\n"
        + f"User {Index_User}'s Eye-gaze Angular Velocity\n"
        + "Descriptive Statistics\n"
        + Data_["EyeGazeAngularVelocity"].describe().__str__()
    )

    print_("\n\n")

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute a graphical category plot of the `EyeGazeAngularVelocity` data.

    ColorMap = seaborn.hls_palette(
        n_colors=6,
        h=0.17 + 0.66 * (Index_User / (len(Indexes_Users) - 1)),
        l=0.50,
        s=0.50,
    )

    Palette_ = seaborn.color_palette(ColorMap)
    Palette_.reverse()

    seaborn.set_theme(
        context="notebook",
        style="whitegrid",
        palette=Palette_,
        font="monospace",
    )

    Plot = seaborn.catplot(
        data=Data_,
        x="EyeGazeAngularVelocity",
        kind="boxen",
        native_scale=True,
    )

    Plot.figure.suptitle(f"User {Index_User}'s Eye-gaze Angular Velocity")
    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_Stats.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute textual descriptive statistics of the `EyeGazeAngularVelocity` vs `SicknessCategory == 0` data.

    print_(
        "\n"
        + f"User {Index_User}'s Eye-gaze Angular Velocity (Without-sickness)\n"
        + "Descriptive Statistics\n"
        + Data_WithoutSickness["EyeGazeAngularVelocity"].describe().__str__()
    )

    print_("\n\n")

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute textual descriptive statistics of the `EyeGazeAngularVelocity` vs `SicknessCategory == 1` data.

    print_(
        "\n"
        + f"User {Index_User}'s Eye-gaze Angular Velocity (With-sickness)\n"
        + "Descriptive Statistics\n"
        + Data_WithSickness["EyeGazeAngularVelocity"].describe().__str__()
    )

    print_("\n\n")

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute textual summary of compare means statistical tests on `"EyeGazeAngularVelocity"` items with `SicknessCategory == "0"` vs `SicknessCategory == "1"`.

    CompareMeans = StatsModels_WeightStats.CompareMeans(
        StatsModels_WeightStats.DescrStatsW(Data_WithoutSickness["EyeGazeAngularVelocity"]),
        StatsModels_WeightStats.DescrStatsW(Data_WithSickness["EyeGazeAngularVelocity"]),
    )

    print_(
        "\n"
        + f"User {Index_User}'s Eye-gaze Angular Velocity\n"
        + "[With-sickness vs. Without-sickness]\n"
        + "Compare Means Statistical Tests\n"

        + CompareMeans.summary(
            use_t=True,
            alpha=0.05,
            usevar="unequal",
            value=0.0
        ).__str__()
    )

    print_("\n\n")

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute a graphical category plot of the `EyeGazeAngularVelocity` vs `SicknessCategory` data.

    Plot = seaborn.catplot(
        data=Data_,
        x="EyeGazeAngularVelocity",
        y="SicknessCategory",
        hue="SicknessCategory",
        kind="boxen",
        order=[1, 0],
        native_scale=True,
    )

    Plot.figure.suptitle(f"User {Index_User}'s Eye-gaze Angular Velocity vs. Sickness")
    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_VsSick_Stats.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute a graphical line plot of the `Time` vs. `EyeGazeAngularVelocity` comprehensive (with-sickness and without-sickness) data.

    Plot = seaborn.relplot(
        data=Data_,
        x="Time",
        y="EyeGazeAngularVelocity",
        size=0,
        sizes=[0.2],
        kind="line",
    )

    Plot.figure.suptitle(f"User {Index_User}'s Comprehensive Eye-gaze Angular Velocity Timeline")
    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_VsTime_Comp.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute a graphical line plot of the `Time` vs. `EyeGazeAngularVelocity` with-sickness data.

    Plot = seaborn.relplot(
        data=Data_WithSickness,
        x="Time",
        y="EyeGazeAngularVelocity",
        hue=0,
        palette=[Palette_[1]],
        size=0,
        sizes=[0.2],
        kind="line",
    )

    Plot.figure.suptitle(f"User {Index_User}'s With-sickness Eye-gaze Angular Velocity Timeline")
    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_VsTime_WSick.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute a graphical line plot of the `Time` vs. `EyeGazeAngularVelocity` without-sickness data.

    Data_WithoutSickness = Data_WithoutSickness

    Plot = seaborn.relplot(
        data=Data_WithoutSickness,
        x="Time",
        y="EyeGazeAngularVelocity",
        hue=0,
        palette=[Palette_[2]],
        size=0,
        sizes=[0.2],
        kind="line",
    )

    Plot.figure.suptitle(f"User {Index_User}'s Without-sickness Eye-gaze Angular Velocity Timeline")
    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_VsTime_WOSick.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute textual and graphical least square spectual analyses info of the `Time` vs. `EyeGazeAngularVelocity` comprehensive (with-sickness and without-sickness) data. We can deem the least square spectual analyses as a kind of discrete Fourier transform analyses designed for time-series data with uneven time-steps.

    # Compute textual frequency-amplitude-phase info.

    Frequencies = numpy.linspace(
        start=1.0,
        stop=32.0,
        num=32 + 0,
        endpoint=True,
    )

    Periodogram = signal.lombscargle(
        x=Data_["Time"],
        y=Data_["EyeGazeAngularVelocity"],
        freqs=Frequencies,
        normalize="amplitude",
        floating_mean=True,
    )

    Amplitudes = []
    Phases = []

    for Index, (Frequency, Period) in enumerate(zip(Frequencies, Periodogram)):
        Amplitude = numpy.abs(Period)
        Phase = numpy.angle(Period, deg=True)
        Amplitudes.append(Amplitude)
        Phases.append(Phase)
    # end for

    print_(
        f"User {Index_User}'s Comprehensive Eye-gaze Angular Velocity\n"
        + "Least Square Spectral Analyses"
    )

    print_("Frequency\t\tAmplitude\t\tPhase")

    for Index, (Frequency, Amplitude, Phase) in enumerate(zip(Frequencies, Amplitudes, Phases)):
        print_(f"{Frequency:.6f}\t\t{Amplitude:.6f}\t\t{Phase:.6f}")
    # end for

    print_("\n\n")

    # Compute graphical frequency-amplitude info.

    Data_Spectrum = pandas.DataFrame({
        "Frequency": Frequencies,
        "Amplitude": Amplitudes,
        "Phase": Phases,
    })

    Plot = seaborn.relplot(
        data=Data_Spectrum,
        x="Frequency",
        y="Amplitude",
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s Comprehensive Eye-gaze Angular Velocity\n"
        + "[Frequency - Amplitude] Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqAmp_Comp.png"),
        dpi=Plot.figure.dpi,
    )

    # Compute graphical frequency-phase info.

    Plot = seaborn.relplot(
        data=Data_Spectrum,
        x="Frequency",
        y="Phase",
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s Comprehensive Eye-gaze Angular Velocity\n"
        + "[Frequency - Phase] Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqPh_Comp.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute textual and graphical least square spectual analyses info of the `Time` vs. `EyeGazeAngularVelocity` with-sickness data. We can deem the least square spectual analyses as a kind of discrete Fourier transform analyses designed for time-series data with uneven time-steps. 

    # Compute textual frequency-amplitude-phase info.

    Frequencies = numpy.linspace(
        start=1.0,
        stop=32.0,
        num=32 + 0,
        endpoint=True,
    )

    Periodogram = signal.lombscargle(
        x=Data_WithSickness["Time"],
        y=Data_WithSickness["EyeGazeAngularVelocity"],
        freqs=Frequencies,
        normalize="amplitude",
        floating_mean=True,
    )

    Amplitudes = []
    Phases = []

    for Index, (Frequency, Period) in enumerate(zip(Frequencies, Periodogram)):
        Amplitude = numpy.abs(Period)
        Phase = numpy.angle(Period, deg=True)
        Amplitudes.append(Amplitude)
        Phases.append(Phase)
    # end for

    print_(
        f"User {Index_User}'s With-sickness Eye-gaze Angular Velocity\n"
        + "Least Square Spectral Analyses"
    )

    print_("Frequency\t\tAmplitude\t\tPhase")

    for Index, (Frequency, Amplitude, Phase) in enumerate(zip(Frequencies, Amplitudes, Phases)):
        print_(f"{Frequency:.6f}\t\t{Amplitude:.6f}\t\t{Phase:.6f}")
    # end for

    print_("\n\n")

    # Compute graphical frequency-amplitude info.

    Data_WithSickness_Spectrum = pandas.DataFrame({
        "Frequency": Frequencies,
        "Amplitude": Amplitudes,
        "Phase": Phases,
    })

    Plot = seaborn.relplot(
        data=Data_WithSickness_Spectrum,
        x="Frequency",
        y="Amplitude",
        hue=0,
        palette=[Palette_[1]],
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s With-sickness Eye-gaze Angular Velocity\n"
        + "[Frequency - Amplitude] Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqAmp_WSick.png"),
        dpi=Plot.figure.dpi,
    )

    # Compute graphical frequency-phase info.

    Plot = seaborn.relplot(
        data=Data_WithSickness_Spectrum,
        x="Frequency",
        y="Phase",
        hue=0,
        palette=[Palette_[1]],
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s With-sickness Eye-gaze Angular Velocity\n"
        + "[Frequency - Phase] Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqPh_WSick.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ====
    # - Compute textual and graphical least square spectual analyses info of the `Time` vs. `EyeGazeAngularVelocity` without-sickness data. We can deem the least square spectual analyses as a kind of discrete Fourier transform analyses designed for time-series data with uneven time-steps.
    
    # Compute textual frequency-amplitude-phase info.

    Frequencies = numpy.linspace(
        start=1.0,
        stop=32.0,
        num=32 + 0,
        endpoint=True,
    )

    Periodogram = signal.lombscargle(
        x=Data_WithoutSickness["Time"],
        y=Data_WithoutSickness["EyeGazeAngularVelocity"],
        freqs=Frequencies,
        normalize="amplitude",
        floating_mean=True,
    )

    Amplitudes = []
    Phases = []

    for Index, (Frequency, Period) in enumerate(zip(Frequencies, Periodogram)):
        Amplitude = numpy.abs(Period)
        Phase = numpy.angle(Period, deg=True)
        Amplitudes.append(Amplitude)
        Phases.append(Phase)
    # end for

    print_(
        f"User {Index_User}'s Without-sickness Eye-gaze Angular Velocity\n"
        + "Least Square Spectral Analyses"
    )

    print_("Frequency\t\tAmplitude\t\tPhase")

    for Index, (Frequency, Amplitude, Phase) in enumerate(zip(Frequencies, Amplitudes, Phases)):
        print_(f"{Frequency:.6f}\t\t{Amplitude:.6f}\t\t{Phase:.6f}")
    # end for

    print_("\n\n")

    # Compute graphical frequency-amplitude info.

    Data_WithoutSickness_Spectrum = pandas.DataFrame({
        "Frequency": Frequencies,
        "Amplitude": Amplitudes,
        "Phase": Phases,
    })

    Plot = seaborn.relplot(
        data=Data_WithoutSickness_Spectrum,
        x="Frequency",
        y="Amplitude",
        hue=0,
        palette=[Palette_[2]],
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s Without-sickness Eye-gaze Angular Velocity\n"
        + "[Frequency - Amplitude] Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqAmp_WOSick.png"),
        dpi=Plot.figure.dpi,
    )

    # Compute graphical frequency-phase info.

    Plot = seaborn.relplot(
        data=Data_WithoutSickness_Spectrum,
        x="Frequency",
        y="Phase",
        hue=0,
        palette=[Palette_[2]],
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s Without-sickness Eye-gaze Angular Velocity\n"
        + "[Frequency - Phase] Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqPh_WOSick.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute textual and graphical info of the differences between the comprehensive and with-sickness frequency spectrums.
    
    # Compute textual frequency-amplitude-phase differences info.

    Data_Spectrum_Diff = pandas.DataFrame({
        "Frequency": Data_Spectrum["Frequency"],
        "Difference_InAmplitude": numpy.abs(Data_WithSickness_Spectrum["Amplitude"] - Data_Spectrum["Amplitude"]),
        "Difference_InPhase": numpy.abs(Data_WithSickness_Spectrum["Phase"] - Data_Spectrum["Phase"]),
    })

    print_(
        f"User {Index_User}'s Eye-gaze Angular Velocity\n"
        + "Comprehensive vs. With-Sickness\n"
        + "Differences in Least Square Spectral Analyses"
    )

    print_("Frequency\t\tDifference_InAmplitude\t\tDifference_InPhase")

    for Index, (Frequency, Diff_InAmplitude, Diff_InPhase) \
        in enumerate(zip(
            Data_Spectrum_Diff["Frequency"],
            Data_Spectrum_Diff["Difference_InAmplitude"],
            Data_Spectrum_Diff["Difference_InPhase"],
        )):
        print_(f"{Frequency:.6f}\t\t{Diff_InAmplitude:.6f}\t\t\t{Diff_InPhase:.6f}")
    # end for

    print_("\n\n")

    # Compute graphical frequency-amplitude differences info.

    Plot = seaborn.relplot(
        data=Data_Spectrum_Diff,
        x="Frequency",
        y="Difference_InAmplitude",
        hue=0,
        palette=[Palette_[3]],
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s Eye-gaze Angular Velocity\n"
        + "Comprehensive vs. With-Sickness\n"
        + "[Frequency - Amplitude] Difference Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqAmpDiff_Comp_VsWSick.png"),
        dpi=Plot.figure.dpi,
    )

    # Compute graphical frequency-phase differences info.

    Plot = seaborn.relplot(
        data=Data_Spectrum_Diff,
        x="Frequency",
        y="Difference_InPhase",
        hue=0,
        palette=[Palette_[3]],
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s Eye-gaze Angular Velocity\n"
        + "Comprehensive vs. With-Sickness\n"
        + "[Frequency - Phase] Difference Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqPhDiff_Comp_VsWSick.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Compute textual and graphical info of the differences between the with-sickness and without-sickness frequency spectrums.
    
    # Compute textual frequency-amplitude-phase differences info.

    Data_Spectrum_Diff_WSickVSWOSick = pandas.DataFrame({
        "Frequency": Data_Spectrum["Frequency"],
        
        "Difference_InAmplitude"\
        : numpy.abs(Data_WithSickness_Spectrum["Amplitude"] - Data_WithoutSickness_Spectrum["Amplitude"]),
        
        "Difference_InPhase"\
        : numpy.abs(Data_WithSickness_Spectrum["Phase"] - Data_WithoutSickness_Spectrum["Phase"]),
    })

    print_(
        f"User {Index_User}'s Eye-gaze Angular Velocity\n"
        + "With-Sickness vs. Without-Sickness\n"
        + "Differences in Least Square Spectral Analyses"
    )

    print_("Frequency\t\tDifference_InAmplitude\t\tDifference_InPhase")

    for Index, (Frequency, Diff_InAmplitude, Diff_InPhase) \
        in enumerate(zip(
            Data_Spectrum_Diff_WSickVSWOSick["Frequency"],
            Data_Spectrum_Diff_WSickVSWOSick["Difference_InAmplitude"],
            Data_Spectrum_Diff_WSickVSWOSick["Difference_InPhase"],
        )):
        print_(f"{Frequency:.6f}\t\t{Diff_InAmplitude:.6f}\t\t\t{Diff_InPhase:.6f}")
    # end for

    print_("\n\n")

    # Compute graphical frequency-amplitude differences info.

    Plot = seaborn.relplot(
        data=Data_Spectrum_Diff_WSickVSWOSick,
        x="Frequency",
        y="Difference_InAmplitude",
        hue=0,
        palette=[Palette_[4]],
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s Eye-gaze Angular Velocity\n"
        + "With-Sickness vs. Without-Sickness\n"
        + "[Frequency - Amplitude] Difference Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqAmpDiff_WSick_VsWOSick.png"),
        dpi=Plot.figure.dpi,
    )

    # Compute graphical frequency-phase differences info.

    Plot = seaborn.relplot(
        data=Data_Spectrum_Diff_WSickVSWOSick,
        x="Frequency",
        y="Difference_InPhase",
        hue=0,
        palette=[Palette_[4]],
        size=0,
        sizes=[1],
        kind="line",
    )

    Plot.figure.suptitle(
        f"User {Index_User}'s Eye-gaze Angular Velocity\n"
        + "With-Sickness vs. Without-Sickness\n"
        + "[Frequency - Phase] Difference Spectrum"
    )

    pyplot.tight_layout(rect=[0, 0, 1, 1])
    pyplot.show()

    Plot.figure.savefig(
        Path_Output_Figure_Find(f"fig_User{Index_User}_EyeGazeAngVel_FreqPhDiff_WSick_VsWOSick.png"),
        dpi=Plot.figure.dpi,
    )

    # ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== ==== 
    # - Complete analyzing the preprocessed data.

    print_(f"End analyzing data for user {Index_User}")
    print_("\n\n\n\n")
# end for

File_Output_Text.close()
````
